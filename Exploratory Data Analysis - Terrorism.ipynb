```python
killdata=Data.loc[:,'nkill']
print('No of People Killed by terror attack ',killdata.dropna().sum())
```

```python
# Number of people killed in terror attack is 411868.0
```

```python
attackdata=Data.loc[:,'attacktype1':'attacktype1_txt']
attackdata
```

```python
typekilldata=pd.concat([killdata,attackdata],axis=1)
typekilldata
```

```python
typekillformatdata=typekilldata.pivot_table(columns='attacktype1_txt',values='nkill',aggfunc='sum')
typekillformatdata
```

```python
labels=typekillformatdata.columns.tolist()
transpoze=typekillformatdata.T
values=transpoze.values.tolist()
fig,ax=plt.subplots(figsize=(20,20),subplot_kw=dict(aspect='equal'))
plt.pie(values,startangle=90,autopct='%.2f%%',shadow=True)
plt.title('Type Of Terrorist Attack Causes Death')
plt.legend(labels,loc='upper right')
plt.show()
```

```python
# Armed assault and bombing/explosion are seen to be the cause of 77% of the deaths in these attacks. 
# This rate is why these attacks are used so many times in terrorist actions. 
# This is how dangerous weapons and explosives are to the world.
```

```python
countrydata=Data.loc[:,'country':'country_txt']
countrydata
```

```python
countrykilldata=pd.concat([countrydata,killdata],axis=1)
countrykilldata
```

```python
print('No of people killed in countries',countrykilldata.dropna().sum())
```

```python
countrykillformatdata=countrykilldata.pivot_table(columns='country_txt',values='nkill',aggfunc='sum')
countrykillformatdata
```

```python
countrykillformatdata.shape
```

```python
fig_size=plt.rcParams['figure.figsize']
fig_size[0]=25
fig_size[1]=25
plt.rcParams['figure.figsize']=fig_size
```

```python
labels=countrykillformatdata.columns.tolist()
labels=labels[:50]
index=np.arange(len(labels))
transpoze=countrykillformatdata.T
values=transpoze.values.tolist()
values=values[:50]
values=[int(i[0]) for i in values ] # convert float into int
colors = ['red', 'green', 'blue', 'purple', 'yellow', 'brown', 'black', 'gray', 'magenta', 'orange'] 
fig,ax=plt.subplots(1,1)
ax.yaxis.grid(True)
plt.bar(index,values,color=colors,width=0.5)
plt.ylabel('Killed People',fontsize=15)
plt.xticks(index,labels,fontsize=15,rotation=90)
plt.title('Number Of People Killed By Country',fontsize=20)
plt.show()
```

```python
# Here in Afghanistan most number of people were killed.
```
